```python
df["Ram(GBs)"]
```

```python
df["Display(cm)"]=df["Display(cm)"].str.split().str[0]
```

```python
df.head()
```

```python
df["Brand"]=df["Title"].str.split().str[0].str.upper()
```

```python
df.head()
```

```python
df["Processor Brand"]=df["Processor"].str.split().str[0].str.upper()
```

```python
df.head()
```

```python
df["Ram(GBs)"]=pd.to_numeric(df["Ram(GBs)"])
```

```python
df["Storage(GBs)"]=df["Storage(GBs)"].str.split().str[0]
```

```python
df.head()
```

```python
df["Storage(GBs)"]=pd.to_numeric(df["Storage(GBs)"])
```

```python
df["Storage(GBs)"] = df["Storage(GBs)"].apply(lambda x: x * 1024 if x < 16 else x)
```

```python
df.head()
```

```python
df["Ram(GBs)"]=pd.to_numeric(df["Ram(GBs)"])
```

```python
df.head()
```

```python
df["Display(cm)"]=pd.to_numeric(df["Display(cm)"])
```

```python
df.head()
```

```python
def clean_os(text):
    if 'Windows' in str(text):
        return 1
    elif 'Mac' in str(text) or 'OS X' in str(text):
        return 2  # Apple usually commands a premium, keep it separate
    return 0
```

```python
df["Operating System"]=df["Operating System"].apply(clean_os)
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
X=df[["Ram(GBs)"]]
```

```python
y=df["Price"]
```

```python
regressor.fit(X,y)
```

```python
print(f"Intercept is:{regressor.intercept_}")
```

```python
print(f"Coefficient is: {regressor.coef_[0]}")
```

```python
print("Model Trained Successfully!")
print(f"Accuracy Score: {regressor.score(X, y)}")
```
